```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df['quarter'] = melb_df['Date'].dt.quarter
# display(melb_df.info())
# display(melb_df['quarter'].value_counts())

cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_count = 150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```

```python
import pandas as pd
import re

def get_year_release(arg):
    #     "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    #   
    if len(candidates) > 0:
        #    0,
	#    "("  ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #   ,  None
        return None


d = pd.read_csv('data/ratings_movies.csv')

d['year_release'] = d['title'].apply(get_year_release)

# display(d.info())
# 100836-100818

# d[d['year_release'] == 2010].groupby(by=['genres']).mean().sort_values(by='rating')

d.groupby('userId')['genres'].nunique().sort_values(ascending=False)
```

```python
d.groupby('userId')['rating'].agg(
    ['count', 'mean']
).sort_values(['count', 'mean'], ascending=[True, False])
```

```python
mask = d['year_release'] == 2018
grouped = d[mask].groupby('genres')['rating'].agg(
    ['mean', 'count']
)
grouped[grouped['count'] > 10].sort_values(
    by='mean',
    ascending=False
)
```

```python
d['date'] = pd.to_datetime(d['date'])
d['year_rating'] = d['date'].dt.year
pivot = d.pivot_table(
    index='year_rating',
    columns='genres',
    values='rating',
    aggfunc='mean'
)
display(pivot)
```
